```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
```

```python
# Variables endógenas:
# PIB. no es pib percápita porque no vaos a hacer comparaciones entre economía, ni consideramos cambios en la población
# Precio, p
# Tasa de interés, r
```

```python
#Variables exógenas
# Mercado de bienes
C = 160 # Consumo autónomo ## == cambio en las expectativas de los consumidores.
I = 100 # inversión autónoma ##== Cambio de expectativas sobre el recimiento económico

# Oferta
w = 50 # Salario nominal ##== Sindicatos consiguieron un incremento salaria.
ka = 30000 # stock de capital ##== desatre natural disminuye el stock de capital
L = 225 # trabajo ## == Pandemia mata el 25% se la población.
A = 1 # tecnología ## == Una mejora en la tecnología, aumenta la productividad
```

```python
# Parámetros
# Mercado de bienes
c = 0.4 # PMgC este parámetro está entre cero y uno 0<c<1.
b = 1500 # sensibilidad de la inversión a la tasa de interés.

# Mercado dinero
k = 0.40 # sensibilidad de la demanda de dinero a la renta
h = 500 # sensibilidad de la demanda de dinero a cambios en la tasa de interés

# Oferta
alpha = 0.6 # proporción de la utilización del capital en la producción
```

```python
# Variables de política. Variables de decisión del jugador (las que él maneja)
# Instrumentos de política fiscal
G = 400 # Gasto Público. Sugerencia hacer variaciones de 50. el intervalo de 50 a 1000
t = 0.5 # tasa de impuesto a la renta Esta está entre 0 y 1, 0<t<1 losincrementos de 0.1 (un punto porcentual)

# Instrumentos de política monetaria
# Operaciones de mecado abierto: Compra y venta de bonos.
M = 40 # Oferta monetaria. Oferta monetaría. sugerencia hacer variaciones de 5 unidades, hasta llegar a 40.
```

```python
# definimos algunas simplificaciones
gamma = (1-c*(1-t))

q = 1/alpha
u = (1-alpha)/alpha

pi_e = 0 # Iflación esperada
```

```python
def doxn(vars):
    # desempaquetamos las variables de entrada
    y, r, p = vars  

    # Forzar positividad mínima en p y r
    p = max(p, 1e-6)
    r = max(r, 1e-6)
 
    # ecuaciones
    da = gamma*y - C - I - G + b*r
    m  = M/p - k*y + h*r
    oa = y - ((A**q) * ka * ((1-alpha)**u) * ((p/w)**u))

    # devolvemos la lista con las tres ecuaciones
    return [da, m, oa]
```

```python
# se establece un valor inicial para la solución
y, r, p = fsolve(doxn, (510, 0.3, 2.5))
```

```python
# Using this 'comma' notation, the get the three solutions directly.
print('PIB =', y, 'Precio =', p, 'tasa interés = ', r)
# === Todos los valores redondeados a dos decimales.
# === Esto nos calcula el PIB nominal, debemos calcular el PIB real, esto␣hacerlo así: PIB real = PIB/precio
# === Calcular la tasa de crecimiento como la variación porcentual del PIB real
```

```python
# PIB real
PIB_real = y/p
print(PIB_real)
```

```python
# ingreso del gobierno Nominal
Ingreso = t*y
Saldo = Ingreso - G
print(Ingreso)
print(Saldo)
```
